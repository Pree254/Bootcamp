Student performace analysis

```python
marks = [45, 67, 82, 39, 90, 55, 73]
```

```python
## Print each mark
for mark in marks:
    print(marks)
```

```python
## Classify each mark
for mark in marks:
  if mark >= 50:
     print(mark,"pass")
  else:
     print(mark,"fail")
```

```python
## Create a function
def classify_mark(mark):
  if mark>=50:
    return "pass"
  else:
    return "fail"
```

```python
## Use function in a loop
for mark in marks:
  result = classify_mark(mark)
  print(mark, result)
```

```python
## Calculate average mark
total=0
for mark in marks:
  total += mark
average = total/len(marks)
print("Average Mark:",average)
```
